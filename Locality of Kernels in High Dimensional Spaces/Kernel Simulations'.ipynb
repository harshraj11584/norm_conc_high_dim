```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
def dist_func(data,p=2):
#     N data points in d dimensions
#     data.shape = N X d
    print("(N,d)=",data.shape)
    dists = (np.sum(abs(data)**p,axis=1))**(1/p)    
    return dists.reshape(data.shape[0],1)

dist_func(1*np.ones((5,5)),2)
```

```python
def gaussian_kernel(X,dist_f,sigma=1.0):
    vals = np.exp( (-0.5/sigma**2)*(dist_f(X)**2))
    return vals

gaussian_kernel( 1*np.ones((5,5)),lambda X: dist_func(X,2), 1 )                                                   
```

```python
def kernel_vals(r_vals, dim, sigma):
    return np.exp( (-0.5/sigma**2)*(r_vals**2))
```

When Distance Function is the L2 Norm

```python
dims = [ 2, 5, 10, 100 ]
N = int(1e5)

for d in dims:
    sigma=1
    
    data = np.random.multivariate_normal( mean=np.zeros(d), cov=((sigma)**2)*np.eye(d), size=N)
#     print("data.shape",data.shape)
    
    fig,ax = plt.subplots()
    
    distances_data = dist_func(data,2)
#     print("ker_vals_data.shape",ker_vals_data.shape)
    ax.hist(distances_data,bins=1000,density=True,label="distribution of distances")
    
    r_vals = np.linspace(0,100,N)
    kernel_vals_r = kernel_vals(r_vals,dim=d,sigma=sigma) 
    ax.plot(r_vals,kernel_vals_r,label="kernel values")
    
    ax.set_ylim(0,1)
    ax.set_xlim(0,20)
    ax.set_title("dim="+str(d))
    plt.legend()
```

Using the p-Gaussian Kernel 

```python
def p_kernel_vals(r_vals, dim, p, sigma):
    return np.exp( -(r_vals**p)/(sigma**p))

kernel_vals(np.arange(10), 5, 1)
```

Cauchy Kernel 

```python
def kernel_vals(r_vals, dim, sigma):
    return 1/(1+r_vals/sigma**2)
```

```python
dims = [ 2, 5, 10, 100 ]
N = int(1e5)

for d in dims:
    sigma=1
    
    data = np.random.multivariate_normal( mean=np.zeros(d), cov=((sigma)**2)*np.eye(d), size=N)
#     print("data.shape",data.shape)
    
    fig,ax = plt.subplots()
    
    distances_data = dist_func(data,0.5)
#     print("ker_vals_data.shape",ker_vals_data.shape)
    ax.hist(distances_data,bins=1000,density=True,label="distribution of distances")
    
    r_vals = np.linspace(0,100,N)
    kernel_vals_r = kernel_vals(r_vals,dim=d,sigma=sigma) 
    ax.plot(r_vals,kernel_vals_r,label="kernel values")
    
    ax.set_ylim(0,1)
    #ax.set_xlim(0,20)
    ax.set_title("dim="+str(d))
    plt.legend()
```
